```python
df["Temp_year"] = df["Date"].astype(str).str.extract(r'(\d{4})').astype(float) # adding new temporary column to extract 4 digits of year from the column Date
df["Temp_year"]
```

```python
df[['Date', 'Year', 'Temp_year']].sample(20)
```

```python
df['Year'] = df['Year'].fillna(df['Temp_year']) # Fill in the NaN in the Year column with the Temp year
df[['Date', 'Year', 'Temp_year']].sample(20)
```

```python
df_50 = df[~((df['Year'] < 1976) & (df['Temp_year'] < 1976))]
df_50[["Year", "Temp_year"]]
```

```python
df_50['Match_check'] = df_50['Year'] == df_50['Temp_year'] # Checking if value in Year matches value in Temp year
df_50['Match_check']
```

```python
print(df_50['Match_check'].value_counts())
```

```python
df_50_false = df_50[df_50['Match_check'] == False] # showing how many rows don't match
df_50_false [["Year", "Temp_year"]]
```

```python
df_50['Year'] = pd.to_numeric(df_50['Year'], errors='raise')   # turning the columns into numeric value to compare
df_50['Temp_year'] = pd.to_numeric(df_50['Temp_year'], errors='raise')
```

```python
df_50 = df_50[~(((df_50['Year'].isna()) | (df_50['Year'] == 0.0)) &
                 ((df_50['Temp_year'].isna()) | (df_50['Temp_year'] == 0.0)))]   # removing all rows that do not have any year information
df_50
```

```python
df_50['Year'] = np.where(                                   # replacing missing value in Year with extracted value in Temp_year
    ((df_50['Year'].isna()) | (df_50['Year'] == 0.0)) &
    (df_50['Temp_year'].between(1000, 9999)),
    df_50['Temp_year'],
    df_50['Year']
)


df_50[["Year", "Temp_year"]]
```

## Cleaning column Activity

```python
df_50_last_years.head()
```

```python
activities = df_50_last_years['Activity'].str.strip().str.title().unique()
print("Activities:", activities)
print("Total different activities:", len(activities))
```

```python

# --- Normalize text for robust matching ---
s = df_50_last_years['Activity'].astype(str).str.lower()
s_norm = (
    s.str.replace(r'[-_]', ' ', regex=True)   # unify separators
     .str.replace(r'\s+', ' ', regex=True)    # squeeze multiple spaces
     .str.strip()
)

# --- Regex categories (order matters; first match wins) ---
categories = [
    # Surfing & boards (now also matches: Boogie boarding, Foilboarding,
    # and phrasing like "Sitting/Paddling on surfboard")
    ("Surfing",
     r"\b("
     r"body\s*board(ing)?|boogie\s*board(ing)?|body\s*surf(ing)?|"
     r"surf(ing)?|surf[- ]?ski(ing)?|skimboard(ing)?|"
     r"(paddling|sitting)\s+on\s+surfboard|"
     r"paddle ?board(ing)?|stand ?up paddle(ing)?|sup\b|"
     r"kite[- ]?(surf|board)(ing)?|wind ?surf(ing)?|"
     r"foil(?:\s*board(?:ing)?|boarding|ing)?"
     r")\b"),

    # In-water, casual play, generic short labels
    # (added: standalone 'jumping', 'fell into the water', 'kneeling in the water',
    # typo 'swimmingq')
    ("Swimming / Wading",
     r"\b("
     r"swim(ming)?|swimmingq|wading|tread(ing)? ?water|bathing|float(ing)?|"
     r"playing( in (the )?water| in the surf)?|"
     r"sitting in the water|walking( in (the )?water)?|"
     r"standing|squatting in the water|kneeling in the water|jumping"
     r")\b"
     r"|"
     r"\bfell into (the )?water\b"),

    # Diving & snorkeling (accept UK spelling: snorkelling)
    ("Diving / Snorkeling",
     r"\b("
     r"div(e|ing)|snorkel(l)?(ing)?|hookah|scuba|free ?div(ing)?|skin ?div(ing)?|cage ?div(ing)?"
     r")\b"),

    # Fishing & harvesting (typos included: 'Fihing'; shore-foraging covered)
    ("Fishing (incl. Spearfishing)",
     r"\b("
     r"fishing|fihing|fish\b|spearfish(ing)?|spearish(ing)?|"
     r"longline|lobster(ing)?|lobstering|crab(bing)?|shrimp(ing)?|"
     r"squid|prawn|cast(ing)? a net|trawl(er|ing)?|"
     r"kayak ?fish(ing)?|surf ?fish(ing)?|fly ?fish(ing)?|"
     r"crayfish(ing)?|opihi|scallop(ing)?"
     r")\b"),

    # Boats & paddling craft; expeditions; overboard events
    # (added: 'parasailing', standalone 'paddling', explicit whale-towing)
    ("Boating / Paddling",
     r"\b("
     r"kayak(ing)?|canoe(ing)?|row(ing)?|sail(ing)?|yacht|boat|dinghy|jet ?ski(ing)?|"
     r"outrigger|ferry|catamaran|paddleski(ing)?|paddle(?! *board)|paddling|"
     r"parasail(ing)?|expedition"
     r")\b"
     r"|"
     r"\btowing a dead whale out to sea\b"
     r"|"
     r"\boverboard\b"
     r"|"
     r"round[- ]the[- ]world"),

    # Direct shark interaction (feeding/handling)
    # (added: harassing sharks, dragging stranded shark, attempting to free the shark)
    ("Feeding / Handling Sharks",
     r"\b("
     r"feed(ing)?|hand ?feed(ing)?|touch(ing)?|petting|wrangling|moving captive shark|"
     r"grabbing shark|teasing|bait(ing)? sharks|shark feeding|"
     r"harass(ing)? sharks|"
     r"reviv(ing)?.*sedated shark|removing shark from net|"
     r"dragging (a )?stranded shark( into deeper water)?|dragging .*shark|"
     r"attempt(ing)? to free( the)? shark|"
     r"finning the shark|putting hand in shark tank|lasso a shark"
     r")\b"),

    # Rescue / training / lifeguard
    ("Rescue / Training / Lifeguard",
     r"\b("
     r"rescu(ing|e)|lifeguard|training( exercise)?|lifesaving drill|"
     r"military (ocean|sea) training|attempting to rescue|body recovery"
     r")\b"),

    # Filming, research, observation, publicity
    # (added: 'shark week', 'sightseeing')
    ("Research / Filming / Measuring",
     r"\b("
     r"research|film(ing)?|documentary|photo ?shoot|photoshoot|photograph(ing)?|"
     r"measur(ing)?|tag(ging)?|shark tagging|shark watching|sardine run|"
     r"publicity stunt|shark week|sightseeing|conducting research"
     r")\b"),

    # Disasters & accidents (sea/air/collisions & shipwreck narratives)
    ("Disaster / Accidents (Sea/Air)",
     r"\b("
     r"disaster|sank|capsiz(ed|ing)?|shipwreck|air disaster|tsunami|founder(ing)?|"
     r"sea disaster|sinking|wreck|collision|freighter|launch was run down|"
     r"vessel caught fire|boat.*swamped|ferry boat.*(exploded|caught fire|sank)|"
     r"adrift|sinking of the|foundering"
     r")\b"),

    # Keep truly non-informative labels here
    ("Other / Unknown",
     r"\b(unknown|undisclosed|not stated|murder|washing hands|sitting)\b|^\s*$|^\?$"),
]




# --- Apply categories (regex only) ---
conditions = [s_norm.str.contains(pat, regex=True, na=False) for _, pat in categories]
choices    = [cat for cat, _ in categories]
df_50_last_years['Activity Category'] = np.select(conditions, choices, default='Other / Unknown')

# --- Summary counts per category ---
category_counts = (
    df_50_last_years['Activity Category']
    .value_counts(dropna=False)
    .rename_axis('Activity Category')
    .reset_index(name='Count')
    .sort_values('Count', ascending=False)
)
print(category_counts)


```

```python
s = df["Date"].astype(str).str.strip()
s = s.str.replace(r"(?<=\d)(st|nd|rd|th)", "", regex=True)
s = s.str.replace(r"\b\d{3,4}\s*hrs\b", "", regex=True, )
s = s.replace([r"^\s*(not\s*stated|unknown|undetermined|\?|nan)\s*$", ""],
              np.nan, regex=True) # normalize placeholders to NaN
```

```python
parsed = pd.to_datetime(s, errors="coerce", dayfirst=True) # initial parse attempt
```

```python
mname = s.str.extract(r"(?i)\b(jan|feb|mar|apr|may|jun|jul|aug|sep|sept|oct|nov|dec)\b", expand=False)
month_map = {"jan":1,"feb":2,"mar":3,"apr":4,"may":5,"jun":6,
             "jul":7,"aug":8,"sep":9,"sept":9,"oct":10,"nov":11,"dec":12}
mnum = mname.str.lower().map(month_map)

year_series = pd.to_numeric(df.get("Year", np.nan), errors="coerce")
mask = parsed.isna() & mnum.notna() & year_series.notna()
parsed.loc[mask] = pd.to_datetime(
    dict(year=year_series[mask].astype(int), month=mnum[mask].astype(int), day=1),
    errors="coerce"
) # construct dates where month name and year exist
```

```python
def month_to_season(m):
    if pd.isna(m): return None
    if m in [12,1,2]:  return "Winter"
    if m in [3,4,5]:   return "Spring"
    if m in [6,7,8]:   return "Summer"
    return "Autumn"

df["Season"] = df["Date_parsed"].dt.month.map(month_to_season)

df.loc[df["Season"].isna() & mnum.notna(), "Season"] = mnum.map(month_to_season)

df["Season"] = pd.Categorical(df["Season"],
                              categories=["Winter","Spring","Summer","Autumn"],
                              ordered=True) # ordered category (nice for sorting/plots)
```

## Injury severity

```python
# Function to classify injury severity
def classify_injury(text):
    if pd.isna(text):
        return np.nan

    text = text.lower()

    # Severe injuries
    severe_keywords = [
        "fatal", "death", "amputation", "severed", "torn off",
        "massive", "critical", "major", "severe", "life-threatening",
        "lost", "missing limb", "unconscious"
    ]

    # Medium injuries
    medium_keywords = [
        "deep", "laceration", "multiple bites", "serious", "fracture",
        "severe bleeding", "puncture", "bitten", "torn", "damaged"
    ]

    # Low injuries
    low_keywords = [
        "minor", "small", "superficial", "scratch", "abrasion",
        "bruise", "nipped", "no injury", "slight"
    ]

    if any(word in text for word in severe_keywords):
        return "Severe"
    elif any(word in text for word in medium_keywords):
        return "Medium"
    elif any(word in text for word in low_keywords):
        return "Low"
    else:
        return "Unknown"
```

```python
df_50_last_years['Injury_Severity'] = df_50_last_years['Injury'].apply(classify_injury) # classify injury severity
```

```python
# Display injury severity counts and sample data
print(df_50_last_years['Injury_Severity'].value_counts(dropna=False))
display(df_50_last_years[['Date', 'Activity', 'Injury', 'Injury_Severity']].head(10))
```

```python
severity_order = ["Low", "Medium", "Severe", "Unknown"]
df_50_last_years["Injury_Severity"] = (
    df_50_last_years["Injury_Severity"]
    .fillna("Unknown")
    .replace({"minor":"Low","serious":"Medium","major":"Severe"})  # normalize some common terms
    .astype("category")
    .cat.set_categories(severity_order, ordered=True)
)
```

```python
def severity_bar(df, x, top_n=None, title="", show=True):
    data = df.copy()

    # Clean severity and x
    data["Injury_Severity"] = (
        data["Injury_Severity"]
        .fillna("Unknown")
        .astype(str).str.strip().str.title()
    )
    data["Injury_Severity"] = pd.Categorical(
        data["Injury_Severity"],
        categories=severity_order,
        ordered=True
    )

    # Clean the grouping column
    data[x] = data[x].astype(str).str.strip()
    data = data.replace({"nan": np.nan})  # convert "nan" strings to NaN
    data = data.dropna(subset=[x, "Injury_Severity"])

    # Filter to top N categories by total incidents
    if top_n:
        top_vals = data[x].value_counts(dropna=False).nlargest(top_n).index
        data = data[data[x].isin(top_vals)]

    # Aggregate; observed=True avoids injecting unused categories
    g = (
        data.groupby([x, "Injury_Severity"], observed=True)
        .size()
        .reset_index(name="Attack_Count")
    )

    # Order x by total for nicer sorting
    totals = g.groupby(x)["Attack_Count"].sum().sort_values(ascending=False)
    g[x] = pd.Categorical(g[x], categories=list(totals.index), ordered=True)

    fig = px.bar(
        g.sort_values([x, "Injury_Severity"]),
        x=x, y="Attack_Count",
        color="Injury_Severity",
        category_orders={"Injury_Severity": severity_order},
        barmode="stack",
        title=title,
    )
    fig.update_layout(legend_title="Severity", xaxis_title=x, yaxis_title="Number of Attacks")

    if show:
        fig.show()
    return fig
```

### Handle Missing Values: Drop? Keep? Fill?

### Age replacement values per subgroup (Filling Age Missing values)


```python
# Compute median Age for each subgroup (Country, Sex, Activity, Fatal)
median_age_by_group = df.groupby(['Country', 'Sex', 'Activity', 'Fatal Y/N'])['Age'].median()

print("\nMedian Age per (Country, Sex, Activity, Fatal Y/N) subgroup:")
print(median_age_by_group)


```

#### Distribution of Median Age Acros All (Country, Sex, Activity, Fata Y7N) Groups

```python
# -----------------------------
# Histogram of subgroup median Age values
# -----------------------------

import plotly.express as px

# Convert the median values to a simple DataFrame
median_age_df = median_age_by_group.reset_index()
median_age_df.columns = ['Country', 'Sex', 'Activity', 'Fatal Y/N', 'Median_Age']

print("\nPreview of median Age dataframe:")
print(median_age_df.head())

print("\nNumber of subgroup combinations:", len(median_age_df))

# Create histogram
fig = px.histogram(
    median_age_df,
    x='Median_Age',
    nbins=30,
    title='Distribution of Median Age Across All (Country, Sex, Activity, Fatal Y/N) Subgroups',
    labels={'Median_Age': 'Median Age'},
    color_discrete_sequence=['#4C6EF5']
)

fig.update_layout(
    xaxis_title='Median Age',
    yaxis_title='Number of Subgroups',
    title_x=0.5,
    plot_bgcolor='white'
)

print("\nShowing histogram of subgroup median Age:")
fig.show()

```

```python
# Function to fill Age using subgroup median including Fatal Y/N
def fill_age(row):
    # Only try to fill if Age is missing
    if pd.isna(row['Age']):
        try:
            return median_age_by_group.loc[
                row['Country'],
                row['Sex'],
                row['Activity'],
                row['Fatal Y/N']   
            ]
        except KeyError:
            # If the combination (Country, Sex, Activity, Fatal Y/N) does not exist,
            # we keep Age as NaN for now
            return row['Age']
    else:
        return row['Age']

df['Age'] = df.apply(fill_age, axis=1)

```

```python
# Recreate AgeGroup after imputing Age
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[-np.inf, 18, 45, 65, np.inf],
    labels=[
        'Minors (<18)',
        'Young Adults (18-45)',
        'Early Middle Age (45-65)',
        'Late Middle Age & Seniors (>65)'
    ],
    right=False
)
```

     Since the missingness in Age is MAR, and depends mainly on Country, Sex, Activity and also Fatal (Y/N), we must impute Age conditionally.
     To avoid bias, we used the median Age inside each (Country, Sex, Activity, Fatal) subgroup. This guarantees that the imputed values reflect the demographic and behavioral patterns in the dataset.
