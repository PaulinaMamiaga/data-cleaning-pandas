# 3.1. Cheking for Missing Values in Age

**“Checking for Missing Values in the variable to be cleaned, e.g., Age”** is part of data quality assessment, and it’s done to answer 3 key questions:

### 3.1.1. Is the data complete?

> - By answering this we are verifying whether all observations in the variable to be cleaned (e.g.,Age) or column have valid values in all its rows. 
> - If not, how many are missing? Are they missing randomly or following a pattern?

> **→ This tells us how reliable the column is before using it in our analysis.**

### 3.1.2. How serious is the missingness?

- By answering this we calculate: 

- a) The number of missing values, and
- b) The percentage of missing values relative to the dataset size.

 **→ This helps decide what to do next:**

- **If few are missing (e.g., <5%)** → we can drop them safely or keep them as NaN.
- **If many are missing (e.g., >20%)** → we must analyze or impute carefully.
- **When more than 30% (>30%)  of the data in a column is missing,** it is usually recommended to drop that variable. However, the final decision depends on how critical the variable is for the specific objectives of the analysis. 

### 3.1.3. How could this affect our analysis?

- The goal here is to understand the impact of missing data before making decisions.

**For example:**
> - If the variable to be cleaned (eg., Age) is important for our model or analysis (like analyzing shark attacks by age group,), then missing values could distort averages, distributions, and results.
>
> - If missingness is concentrated in certain regions, activities, injury type, or years, it could indicate a systematic issue in data collection, not just random gaps.

      In summary

 **“Checking for Missing Values in a variable to be cleaned (e.g., Age)”** is about understanding data completeness and quality before we transform or analyze it.

      → It helps us:

> - Detect how much data is missing
> - Evaluate whether the missingness is acceptable
> - Decide the right cleaning strategy (keep, drop, impute, or investigate further)

# 3.2. Analyzing the Impact of missing values in Age

**Objective:** Analyzing the Impact of Missing Values in Age

- To understand whether missing Age data affects the reliability of risk analysis and insurance pricing decisions, and 
- To decide the most appropriate way to handle it.

     We explore:

### 3.2.1. The type of missingness & Data Cleaning strategy

- To analize if the missing values occur randomly or follow a specific pattern. Is it random or systematic?

     Missing Completely at Random (MCAR):
- The missing ages happen by chance. They don’t depend on any other variable.
- *Strategy:* We can safely drop or impute them.

      Missing at Random (MAR):
- Missing ages depend on another variable (e.g., Sex, Country, years, seasons, injury, fatal (yes, no) or Activity).
- *Strategy:* We may need to use grouped imputation (e.g., median age per activity).

       Missing Not at Random (MNAR):
- The reason for missing ages is related to the value itself (e.g., older people tend not to report their age).
- *Strategy:* This is more complex. Dropping rows might introduce bias.

### 3.2.2. The "missingness" relationship between Age and other variables such as Country, Activity, Fatal (Y/N), or Sex to lead Clening Strategy Decision

- To evaluate how missing data affects our variables or conclusions
- We analyze whether missing ages distort relationships or statistical summaries, 

       for example:

- Are missing ages linked to more fatal or non-fatal cases?
- Are key distributions (like gender, activity, injury, fatal or country) different when Age is missing?

### 4.3.1 Age replacement values per subgroup (Filling Age Missing values)


````python
# Compute median Age for each subgroup (Country, Sex, Activity, Fatal)
median_age_by_group = df.groupby(['Country', 'Sex', 'Activity', 'Fatal Y/N'])['Age'].median()

print("\nMedian Age per (Country, Sex, Activity, Fatal Y/N) subgroup:")
print(median_age_by_group)


````

#### 4.3.1.1. Distribution of Median Age Acros All (Country, Sex, Activity, Fata Y7N) Groups

````python
# -----------------------------
# Histogram of subgroup median Age values
# -----------------------------

import plotly.express as px

# Convert the median values to a simple DataFrame
median_age_df = median_age_by_group.reset_index()
median_age_df.columns = ['Country', 'Sex', 'Activity', 'Fatal Y/N', 'Median_Age']

print("\nPreview of median Age dataframe:")
print(median_age_df.head())

print("\nNumber of subgroup combinations:", len(median_age_df))

# Create histogram
fig = px.histogram(
    median_age_df,
    x='Median_Age',
    nbins=30,
    title='Distribution of Median Age Across All (Country, Sex, Activity, Fatal Y/N) Subgroups',
    labels={'Median_Age': 'Median Age'},
    color_discrete_sequence=['#4C6EF5']
)

fig.update_layout(
    xaxis_title='Median Age',
    yaxis_title='Number of Subgroups',
    title_x=0.5,
    plot_bgcolor='white'
)

print("\nShowing histogram of subgroup median Age:")
fig.show()

````

````python
# Function to fill Age using subgroup median including Fatal Y/N
def fill_age(row):
    # Only try to fill if Age is missing
    if pd.isna(row['Age']):
        try:
            return median_age_by_group.loc[
                row['Country'],
                row['Sex'],
                row['Activity'],
                row['Fatal Y/N']   
            ]
        except KeyError:
            # If the combination (Country, Sex, Activity, Fatal Y/N) does not exist,
            # we keep Age as NaN for now
            return row['Age']
    else:
        return row['Age']

df['Age'] = df.apply(fill_age, axis=1)

````

````python
# Recreate AgeGroup after imputing Age
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[-np.inf, 18, 45, 65, np.inf],
    labels=[
        'Minors (<18)',
        'Young Adults (18-45)',
        'Early Middle Age (45-65)',
        'Late Middle Age & Seniors (>65)'
    ],
    right=False
)
````

     Since the missingness in Age is MAR, and depends mainly on Country, Sex, Activity and also Fatal (Y/N), we must impute Age conditionally.
     To avoid bias, we used the median Age inside each (Country, Sex, Activity, Fatal) subgroup. This guarantees that the imputed values reflect the demographic and behavioral patterns in the dataset.

#### 4.3.7. Why AgeGroup Shows 15.7% Missing & How to solve This?

     This is not a real missing-value problem!!! It happens because of the order in which the cleaning steps were executed.

- AgeGroup was created before applying KNN Imputation
At that moment, the Age column still had 617 missing values.
>
- Since AgeGroup is derived from Age, those missing values automatically produced:
**Missing Age.** So AgeGroup simply inherited the NaNs that already existed in Age.

         After applying KNN, Age no longer has missing values
         However…

- AgeGroup does NOT update automatically, because it is a derived column (created manually using pd.cut()).
- Therefore, even though Age has no missing values after KNN, the AgeGroup column still contains the old NaNs, which is why the chart still shows 15.7% missing.

      Why Median Imputation + KNN Must Be Done Before Grouping?


‼️Because if we create AgeGroup before imputing Age,
the group labels will be wrong or missing.

     Correct workflow:

- Clean Age → standardize format
>
- Conditional Median Imputation (Country, Sex, Activity, Fatal)
>
- KNN Imputation (fills remaining missing values)
>
- Recreate AgeGroup from the fully imputed Age column


      Only then:

- ✔ Age is complete
- ✔ AgeGroup is complete
- ✔ No artificial missing values appear in your analysis

This ensures that AgeGroup reflects the true distribution after cleaning, and not the incomplete data from earlier steps.

     How to solve this?

We will have to recreate AgeGroup again.
